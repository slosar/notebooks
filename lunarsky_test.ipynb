```python
import numpy as np
import astropy.units as u
from lunarsky import MCMF, SkyCoord, LunarTopo, MoonLocation
```

```python
N=100
phi = np.linspace(0,2*np.pi,N)
```

```python
R=2000*u.km
x,y,z=R*np.cos(phi),R*np.sin(phi),R*np.zeros(N)
sky_coords_units = SkyCoord(MCMF(x,y,z))
```

```python
R=2000
x,y,z=R*np.cos(phi),R*np.sin(phi),R*np.zeros(N)
sky_coords_sans_units = SkyCoord(MCMF(x,y,z))
```

```python
L = MoonLocation.from_selenodetic(lon=180*u.deg, lat=0, height=0)
altaz_units = sky_coords_units.transform_to(LunarTopo(location=L))
altaz_sans_units = sky_coords_sans_units.transform_to(LunarTopo(location=L))
```
