```python
import numpy as np
```

```python
Nfft = 4096
Ntaps = 4
Nboost = 100
Ntot = Nfft*Ntaps*Nboost
x = (np.arange(Ntot) - Ntot//2)/(Nfft);
wfull = np.sinc(x)
wcut = wfull*(np.abs(x) < Ntaps/2)
```

```python
plt.plot(wfull)
plt.plot(wcut)
```

```python
resp_full = np.abs(np.fft.rfft(wfull)**2)
resp_cut = np.abs(np.fft.rfft(wcut)**2)
```

```python
plt.plot(resp_full[:500])
plt.plot(resp_cut[:500])
```
